```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df['Flower name'] = iris_df.Target.apply(lambda x: iris.target_names[x])
iris_df.head()
```

```python
set_df = iris_df[:50]
versi_df = iris_df[50:100]
virgi_df = iris_df[100:150]
```

```python
set_df.head()
```

```python
versi_df.head()
```

```python
plt.scatter(set_df['petal length (cm)'], set_df['petal width (cm)'], color = 'red', marker = '*')
plt.scatter(versi_df['petal length (cm)'], versi_df['petal width (cm)'], color = 'blue', marker = '+')
plt.scatter(virgi_df['petal length (cm)'], set_df['petal width (cm)'], color = 'green', marker = '.')
plt.xlabel('Petal Length')
plt.ylabel('Sepal width')
```

```python
from sklearn.model_selection import train_test_split
```

```python
indep_var = iris_df.drop(['Target', 'Flower name'], axis = 'columns')
dep_var = iris_df.Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(indep_var, dep_var, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.predict([[5.1,	3.5,	1.4,	0.2]])
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
plt.figure(figsize = (7,6))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('True')
```
