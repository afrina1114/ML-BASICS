```python
plt.scatter(insurance_df.Age,insurance_df.gets_insurance, color = 'red',marker = '+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(insurance_df[['Age']],insurance_df.gets_insurance,test_size = 0.1)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2 = LogisticRegression()
```

```python
model2.fit(X_train,y_train)
```

```python
X_test
```

```python
y_predicted = model2.predict(X_test)
```

```python
y_predicted
```

```python
model2.predict_proba(X_test)
```
