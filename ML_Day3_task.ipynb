```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
indep_var = np.array([1,2,3,4,5,6]).reshape(-1,1)
dep_var = np.array([2,4,6,7,10,11])
```

```python
indep_var
```

```python
dep_var
```

```python
model1 = LinearRegression()
```

```python
model1.fit(indep_var,dep_var)
```

```python
indep_var_test = np.array([6,7,8,9,10]).reshape(-1,1)
```

```python
predic_value = model1.predict(indep_var_test)
```

```python
print("The predicted values are:",predic_value)
```

```python
#plt.(plot)(x,y)
plt.scatter(indep_var,dep_var,color ='red',label ='linear line')
```

```python
plt.plot(indep_var,model1.predict(indep_var),color = 'blue',label = 'Linear Regression Model')
```

```python
plt.scatter(indep_var_test,predic_value, color = 'green',label = 'Predicted Value')
```

```python
plt.scatter(indep_var,dep_var)
```

```python
plt.scatter(indep_var,dep_var,color = 'red',label ='linear line')
plt.plot(indep_var,model1.predict(indep_var), color = 'blue',label = 'linear Regression Model')
plt.xlabel('indep_variable')
plt.ylabel('dep_variable')
plt.show()
```
